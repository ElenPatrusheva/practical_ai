```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('data/photo-1.jpg')
```

```python
plt.imshow(img)
```

```python
distCoeff = np.zeros((4,1),np.float64)

# TODO: add your coefficients here!
k1 = -1.0e-5 * 3; # negative to remove barrel distortion
k2 = 0.0;
p1 = 0.0;
p2 = 0.0;

distCoeff[0,0] = k1;
distCoeff[1,0] = k2;
distCoeff[2,0] = p1;
cam = np.eye(3,dtype=np.float32)
cam[0,2] = img.shape[1]/2.0   # define center x
cam[1,2] = img.shape[0]/2.0 # define center y
cam[0,0] = 10.        # define focal length x
cam[1,1] = 10.
```

```python
dst = cv2.undistort(img,cam,distCoeff)
```

```python
plt.imshow(dst)
```
